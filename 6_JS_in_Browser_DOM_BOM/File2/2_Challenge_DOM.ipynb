### ğŸ’» The Solution: Using event.target

```javascript
// Step 1: Select the parent list
document.getElementById("tea-list")
    .addEventListener('click', function(event) {
        
        // Step 2: event.target gives us the CLICKED element!
        console.log(event.target);           // <li class="tea-item">Oolong Tea</li>
        console.log(event.target.textContent); // "Oolong Tea"
        
        // Step 3: Alert which tea was selected
        alert("You selected: " + event.target.textContent);
    });
```

### ğŸ›¡ï¸ Adding Safety: The matches() Method

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #d29922; margin-top: 15px;">
  <strong style="color: #d29922;">âš ï¸ THE EDGE CASE:</strong><br><br>
  <span style="color: #c9d1d9;">What if user clicks on the <code>&lt;ul&gt;</code> itself (the gap between items)? Or what if there are other elements inside the list that aren't tea items? We need a <strong>safety check!</strong></span>
</div>

```javascript
document.getElementById("tea-list")
    .addEventListener('click', function(event) {
        
        // Safety check: Only proceed if clicked element matches our selector!
        if (event.target.matches(".tea-item")) {
            alert("You selected: " + event.target.textContent);
        }
        // If user clicked elsewhere (gap, other element), nothing happens
    });
```

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff; margin-top: 15px;">
  <strong style="color: #58a6ff;">ğŸ“š CONCEPT: element.matches(selector)</strong><br><br>
  <span style="color: #c9d1d9;">The <code>matches()</code> method checks if an element matches a CSS selector. Returns <code>true</code> or <code>false</code>.</span>
  <ul style="color: #c9d1d9; margin-top: 10px;">
    <li><code>element.matches(".class-name")</code> â€” Check for class</li>
    <li><code>element.matches("#id-name")</code> â€” Check for ID</li>
    <li><code>element.matches("li")</code> â€” Check for tag</li>
  </ul>

```javascript
// Step 1: Select the form (not the button!)
document.getElementById("feedback-form")
    // Step 2: Listen for 'submit' event (not 'click'!)
    .addEventListener('submit', function(event) {
        
        // Step 3: PREVENT the default form submission!
        event.preventDefault();
        
        // Step 4: Get the input value using .value
        let feedback = document.getElementById("feedback-input").value;
        
        // Step 5: Display the feedback
        document.getElementById("feedback-display").textContent = 
            `Feedback is: ${feedback}`;
    });
```

### ğŸ’» The Solution

```javascript
// Notice: We add the listener directly to 'document'
// (not getElementById â€” document itself!)

document.addEventListener('DOMContentLoaded', function() {
    
    // This code runs ONLY after DOM is fully loaded
    document.getElementById("dom-status").textContent = "DOM fully loaded!";
    
    // Now it's safe to:
    // - Access any element
    // - Add event listeners
    // - Fetch data from APIs
    // - etc.
});
```

### ğŸ” Example: Grabbing All Links on a Page

```javascript
// Get ALL anchor tags on the page
let anchorElements = document.getElementsByTagName("a");

console.log(anchorElements);        // HTMLCollection(138)
console.log(anchorElements.length); // 138 links!

// Access individual elements like an array:
console.log(anchorElements[0].textContent); // First link's text
```

### ğŸ”„ Looping Through HTML Collections

```javascript
let anchorElements = document.getElementsByTagName("a");

// Classic for loop works!
for (let i = 0; i < 5; i++) {  // Just first 5
    if (anchorElements[i]) {
        console.log(anchorElements[i].textContent);
    }
}

// You can even CHANGE them:
anchorElements[0].textContent = "I changed this link!";
```
