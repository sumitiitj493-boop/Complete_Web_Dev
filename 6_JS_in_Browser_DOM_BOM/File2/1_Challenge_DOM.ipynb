```javascript
// Step 1: Grab button, add click listener
document.getElementById("add-new-item")
    .addEventListener('click', function() {
        
        // Step 2: Grab the shopping list (parent)
        let shoppingList = document.getElementById("shopping-list");
        
        // Step 3: CREATE a new element (exists in memory, NOT in DOM yet!)
        let newItem = document.createElement("li");
        
        // Step 4: Add content to the new element
        newItem.textContent = "Eggs";
        
        // Step 5: APPEND the new element to the parent (NOW it's in the DOM!)
        shoppingList.appendChild(newItem);
    });
```
