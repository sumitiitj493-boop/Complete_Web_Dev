## 4.2 Navigator Object ğŸ•µï¸

```javascript
// Browser and device information
console.log(navigator.userAgent);
// "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/120.0.0.0"

console.log(navigator.language);      // "en-US"
console.log(navigator.cookieEnabled); // true
console.log(navigator.onLine);        // true (internet connected)
console.log(navigator.platform);      // "Win32"

// Geolocation API
navigator.geolocation.getCurrentPosition(
    position => console.log(position.coords.latitude)
);
```

## 4.3 Screen Object ğŸ–¥ï¸

```javascript
// Screen properties
console.log(screen.width);        // 1920 (screen width)
console.log(screen.height);       // 1080 (screen height)
console.log(screen.availWidth);   // 1920 (available width)
console.log(screen.availHeight);  // 1040 (minus taskbar)
console.log(screen.colorDepth);   // 24 (bits per pixel)
```

## 4.4 History Object â®ï¸

```javascript
// Navigate browser history
history.back();     // Go back one page
history.forward();  // Go forward one page
history.go(-2);     // Go back 2 pages
history.go(1);      // Go forward 1 page

console.log(history.length);  // Number of entries
```

```html
                            document
                                â”‚
                                â”‚
                              <html>
                                â”‚
                â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
                â”‚                               â”‚
             <head>                          <body>
                â”‚                              â”‚
        â”Œâ”€â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”         â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
        â”‚                â”‚         â”‚           â”‚           â”‚           â”‚
     <title>          <meta>     <h1>        <p>        <div>      <button>
        â”‚                                                 â”‚
   "Page Title"                                   â”Œâ”€â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”
                                                  â”‚                â”‚
                                                 <span>          <img>

Key Relationships:
â”œâ”€ Parent: The containing element
â”œâ”€ Children: Elements directly inside
â”œâ”€ Siblings: Elements at the same level
â””â”€ Descendants: All nested elements below


## DOM Traversal Properties

| Property | Returns | Example |
|----------|---------|----------|
| `parentElement` | Parent node | `child.parentElement` |
| `children` | Child elements (HTMLCollection) | `parent.children` |
| `firstElementChild` | First child | `parent.firstElementChild` |
| `lastElementChild` | Last child | `parent.lastElementChild` |
| `nextElementSibling` | Next sibling | `element.nextElementSibling` |
| `previousElementSibling` | Previous sibling | `element.previousElementSibling` |

```javascript
// Example HTML:
// <ul id="list">
//   <li>First</li>
//   <li id="middle">Middle</li>
//   <li>Last</li>
// </ul>

const list = document.getElementById('list');
const middle = document.getElementById('middle');

console.log(list.children);              // [li, li, li]
console.log(list.firstElementChild);     // <li>First</li>
console.log(list.lastElementChild);      // <li>Last</li>

console.log(middle.parentElement);       // <ul id="list">...
console.log(middle.previousElementSibling); // <li>First</li>
console.log(middle.nextElementSibling);     // <li>Last</li>
```

> ğŸ’¡ **Tip:** Use `Element` versions (e.g., `firstElementChild`) instead of `Node` versions (e.g., `firstChild`) to avoid text nodes and comments.

## Function vs Arrow Function - CRITICAL DIFFERENCE!

```javascript
const button = document.querySelector("#myButton");

// Regular function - 'this' refers to the button
button.addEventListener("click", function() {
    console.log(this);  // <button id="myButton">Click Me</button>
    this.style.backgroundColor = "blue";  // Works!
});

// Arrow function - 'this' refers to window
button.addEventListener("click", () => {
    console.log(this);  // Window {...}
    this.style.backgroundColor = "blue";  // Error! Window has no style
});

// Solution: Use event.target with arrow functions
button.addEventListener("click", (event) => {
    console.log(event.target);  // <button id="myButton">Click Me</button>
    event.target.style.backgroundColor = "blue";  // Works!
});
```

## innerHTML vs textContent vs innerText

```javascript
const div = document.querySelector("#myDiv");

// innerHTML - Parses HTML
div.innerHTML = "<strong>Bold Text</strong>";
// Result: Bold Text (actually bold)

// textContent - Plain text only (SAFE from XSS)
div.textContent = "<strong>Bold Text</strong>";
// Result: <strong>Bold Text</strong> (shows literally)

// innerText - Considers CSS (slower)
div.innerText = "Text with line\nbreak";
// Result: Respects display:none, visibility
```

| Property | Parses HTML | Considers CSS | Speed | Use Case |
|----------|-------------|---------------|-------|----------|
| `innerHTML` | âœ… Yes | âŒ No | Fast | Inserting HTML |
| `textContent` | âŒ No | âŒ No | Fastest | Plain text (safe) |
| `innerText` | âŒ No | âœ… Yes | Slowest | Visible text only |

## Working with Attributes

```javascript
const img = document.querySelector("img");

// Get attribute
const src = img.getAttribute("src");
console.log(src);  // "image.jpg"

// Set attribute
img.setAttribute("alt", "A beautiful image");
img.setAttribute("width", "500");

// Remove attribute
img.removeAttribute("width");

// Check if has attribute
if (img.hasAttribute("alt")) {
    console.log("Image has alt text!");
}

// Direct property access (common attributes)
img.src = "newimage.jpg";
img.alt = "New description";

// Data attributes
const btn = document.querySelector("button");
// <button data-user-id="123" data-role="admin">Click</button>

console.log(btn.dataset.userId);  // "123"
console.log(btn.dataset.role);    // "admin"
btn.dataset.active = "true";      // Add data-active="true"
```

## insertBefore, insertAdjacentHTML

```javascript
const list = document.getElementById("myList");

// appendChild - adds to end
const newItem = document.createElement("li");
newItem.textContent = "Last item";
list.appendChild(newItem);

// insertBefore - adds before reference
const firstItem = list.firstElementChild;
const beforeItem = document.createElement("li");
beforeItem.textContent = "Before first";
list.insertBefore(beforeItem, firstItem);

// insertAdjacentHTML - powerful insertion
list.insertAdjacentHTML("beforebegin", "<p>Before list</p>");
list.insertAdjacentHTML("afterbegin", "<li>First in list</li>");
list.insertAdjacentHTML("beforeend", "<li>Last in list</li>");
list.insertAdjacentHTML("afterend", "<p>After list</p>");
```

# 11. Quick Reference Guide ğŸ“‹

Your cheat sheet for DOM manipulation!

## Selecting Elements

```javascript
// By ID (fastest)
document.getElementById("id")

// By Class (live)
document.getElementsByClassName("class")

// By Tag (live)
document.getElementsByTagName("div")

// CSS Selector (first match)
document.querySelector(".class")
document.querySelector("#id")
document.querySelector("div.class")

// CSS Selector (all matches)
document.querySelectorAll(".class")
```

## Modifying Content

```javascript
element.textContent = "text"        // Plain text (safe)
element.innerHTML = "<b>HTML</b>"   // Parse HTML (careful!)
element.innerText = "text"          // Visible text only
```

## Modifying Styles

```javascript
// Inline styles
element.style.color = "red"
element.style.backgroundColor = "blue"
element.style.fontSize = "20px"

// CSS classes (better!)
element.classList.add("class")
element.classList.remove("class")
element.classList.toggle("class")
element.classList.contains("class")
```

## Creating & Removing Elements

```javascript
// Create
const div = document.createElement("div")
div.textContent = "Hello"
parent.appendChild(div)

// Insert
parent.insertBefore(newChild, referenceChild)
element.insertAdjacentHTML("beforeend", "<p>Text</p>")

// Remove
element.remove()
parent.removeChild(child)
```

## Event Handling

```javascript
// Add event listener
element.addEventListener("click", function(event) {
    // event.target - element that triggered
    // event.currentTarget - element listener is on
    // this - element (in regular functions)
})

// Remove event listener
element.removeEventListener("click", functionName)

// Prevent default
event.preventDefault()

// Stop propagation
event.stopPropagation()
```

## DOM Traversal

```javascript
element.parentElement           // Parent
element.children                // Child elements
element.firstElementChild       // First child
element.lastElementChild        // Last child
element.nextElementSibling      // Next sibling
element.previousElementSibling  // Previous sibling
```

# 12. Memory Tricks & One-Liners ğŸ§

    <li><strong>Window:</strong> "The Universe" - Contains everything (DOM + BOM + JS)</li>
    <li><strong>DOM vs HTML:</strong> "Memory vs File" - DOM is the in-memory tree, HTML is the file</li>
    <li><strong>querySelector:</strong> "CSS for JavaScript" - Use CSS selectors to find elements</li>
    <li><strong>Event Delegation:</strong> "One Parent, Many Children" - Parent handles all child events</li>
    <li><strong>preventDefault():</strong> "Stop the Default Dance" - Prevents browser's default action</li>
    <li><strong>this vs arrow:</strong> "Regular = Element, Arrow = Window" - In event handlers</li>
    <li><strong>classList.toggle:</strong> "Smart Switch" - Adds if missing, removes if present</li>
    <li><strong>textContent:</strong> "The Safe Choice" - No HTML parsing, no XSS attacks</li>

## One-Liners to Impress Interviewers

| Question | One-Liner Answer |
|----------|------------------|
| **What is the DOM?** | The in-memory tree representation of HTML that JavaScript can manipulate |
| **DOM vs BOM?** | DOM is the document tree (HTML elements), BOM is browser APIs (navigator, location) |
| **querySelector vs getElementById?** | querySelector uses CSS selectors and returns first match; getElementById is faster but only finds by ID |
| **Event bubbling?** | Events propagate from target up to root, allowing parent elements to handle child events |
| **Why use event delegation?** | Better performance, works with dynamic content, fewer event listeners |
| **textContent vs innerHTML?** | textContent is safe plain text; innerHTML parses HTML and can cause XSS |
| **this in arrow functions?** | Arrow functions inherit this from parent scope; regular functions have this = target element |
| **HTMLCollection vs NodeList?** | HTMLCollection is live and updates automatically; NodeList is usually static |
| **When to use DOMContentLoaded?** | When script is in head without defer and needs to access DOM elements |
| **classList vs className?** | classList has methods (add, remove, toggle); className is just a string |

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Todo App - DOM Mastery</title>
    <style>
        * { margin: 0; padding: 0; box-sizing: border-box; }
        body {
            font-family: Arial, sans-serif;
            background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            min-height: 100vh;
            padding: 40px 20px;
        }
        .container {
            max-width: 600px;
            margin: 0 auto;
            background: white;
            padding: 30px;
            border-radius: 15px;
            box-shadow: 0 20px 60px rgba(0,0,0,0.3);
        }
        h1 {
            color: #667eea;
            text-align: center;
            margin-bottom: 30px;
        }
        .input-group {
            display: flex;
            gap: 10px;
            margin-bottom: 20px;
        }
        input {
            flex: 1;
            padding: 15px;
            border: 2px solid #e0e0e0;
            border-radius: 8px;
            font-size: 16px;
        }
        button {
            padding: 15px 30px;
            background: #667eea;
            color: white;
            border: none;
            border-radius: 8px;
            cursor: pointer;
            font-size: 16px;
            transition: 0.3s;
        }
        button:hover { background: #5568d3; }
        ul {
            list-style: none;
        }
        li {
            padding: 15px;
            background: #f5f5f5;
            margin-bottom: 10px;
            border-radius: 8px;
            display: flex;
            justify-content: space-between;
            align-items: center;
            transition: 0.3s;
        }
        li:hover { background: #e8e8e8; }
        li.completed {
            opacity: 0.6;
            text-decoration: line-through;
        }
        .delete-btn {
            background: #e74c3c;
            padding: 8px 15px;
            font-size: 14px;
        }
        .delete-btn:hover { background: #c0392b; }
    </style>
</head>
<body>
    <div class="container">
        <h1>ğŸ“ My Todo List</h1>
        
        <form id="todoForm" class="input-group">
            <input 
                type="text" 
                id="todoInput" 
                placeholder="Enter a new task..."
                required
            >
            <button type="submit">Add</button>
        </form>
        
        <ul id="todoList"></ul>
    </div>

    <script>
        // Wait for DOM to be ready
        document.addEventListener('DOMContentLoaded', function() {
            
            // Get elements
            const form = document.getElementById('todoForm');
            const input = document.getElementById('todoInput');
            const todoList = document.getElementById('todoList');
            
            // Handle form submission
            form.addEventListener('submit', function(event) {
                event.preventDefault(); // Prevent page refresh
                
                const todoText = input.value.trim();
                
                if (todoText === '') return; // Don't add empty todos
                
                // Create new todo item
                addTodo(todoText);
                
                // Clear input
                input.value = '';
                input.focus();
            });
            
            // Event delegation for list items
            todoList.addEventListener('click', function(event) {
                
                // Handle delete button clicks
                if (event.target.matches('.delete-btn')) {
                    const li = event.target.parentElement;
                    li.remove();
                }
                
                // Handle todo item clicks (toggle complete)
                else if (event.target.matches('li')) {
                    event.target.classList.toggle('completed');
                }
            });
            
            // Function to add a new todo
            function addTodo(text) {
                // Create list item
                const li = document.createElement('li');
                li.textContent = text;
                
                // Create delete button
                const deleteBtn = document.createElement('button');
                deleteBtn.textContent = 'Delete';
                deleteBtn.className = 'delete-btn';
                
                // Append button to list item
                li.appendChild(deleteBtn);
                
                // Append list item to list
                todoList.appendChild(li);
            }
            
            // Add some sample todos
            addTodo('Learn DOM Manipulation');
            addTodo('Build a Todo App');
            addTodo('Master JavaScript');
        });
    </script>
</body>
</html>
```

| `event.preventDefault()` | Stop form submission | Prevents page refresh on submit |
| `input.value.trim()` | Clean input | Removes whitespace from user input |
| `document.createElement('li')` | Create element | Build new DOM nodes programmatically |
| `li.appendChild(deleteBtn)` | Append child | Add elements to the DOM tree |
| Event delegation on `todoList` | Efficient events | One listener handles all todos (including future ones) |
| `event.target.matches('.delete-btn')` | Target checking | Identify which element triggered event |
| `classList.toggle('completed')` | CSS toggle | Switch class on/off with single method |
