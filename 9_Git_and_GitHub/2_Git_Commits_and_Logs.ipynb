# 2. The Staging Area (`git add`) ğŸš‚

The Staging Area is the **"Pre-Screen"** zone. It allows you to pick *exactly* what goes into the next save point.

### The Command
* `git add <filename>`: Stages a specific file.
* `git add .`: Stages **everything** (be careful with this!).

Let's say we only want to save `test1.txt` for now.

# 3. The Commit (`git commit`) ğŸ’¾

Committing moves changes from **Staging** to the **Repository History**.

### The Golden Command
Always use the `-m` flag to provide a message inline.

```bash
$ git commit -m "Add file one"
```

# 4. Viewing History (`git log`) ğŸ“œ

Once you have made commits, you can view the timeline.

### Anatomy of a Log
When you run `git log`, you see detailed info:

<!-- <pre style="white-space: pre; font-family: monospace; overflow-x: auto; color: #c9d1d9;"> -->
```bash

commit 4c8a9...  <--- SHA-1 Hash (Unique ID)
Author: Hitesh <email...>
Date:   Mon Jan 13...

    Add file one  <--- Your Message
```
<!-- </pre> -->

### The Pro Command
For a cleaner view, use `--oneline`.

```bash
# Detailed view
$ git log

# Short view (ID + Message only)
$ git log --oneline

# Output example:
# 4c8a912 Add file one
# a1b2c3d Initial commit
```

# 5. Best Practices: Atomic Commits âš›ï¸

### What is an Atomic Commit?
**One Commit = One Task.**

| âœ… Do This | âŒ Don't Do This |
| :--- | :--- |
| Commit 1: Fix Login Bug | Commit 1: Fix login, add header, delete footer, change colors |
| Commit 2: Add Header | |

### Writing Messages (The Imperative Mood)
Talk to the codebase like you are giving an **Order**.
* âœ… "Add test file" (Present tense, Command)
* âŒ "Added test file" (Past tense)
* âŒ "I added a file because I needed to test" (Too verbose)

# âš¡ Quick Revision & Cheat Sheet

| Command | Description | Memory Trick |
| :--- | :--- | :--- |

| `:q!` | Quits Vim without saving | The **Emergency Exit** |
