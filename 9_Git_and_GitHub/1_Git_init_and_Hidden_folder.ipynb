# 3. Installation & Setup (Windows Edition) ğŸªŸ

### Step 1: Download Git
Go to [git-scm.com](https://git-scm.com) and download the **Windows** installer. Run it and keep hitting "Next" (the defaults are usually fine).

### Step 2: The Terminal
Git is a **Terminal-First** tool. While GUIs exist, the command line gives you superpowers.
On Windows, we recommend using **Git Bash** (which comes with the Git installer) or **PowerShell**.

### Step 3: Verify Installation
Open your terminal and check the version.

```bash
# Check Git version
$ git --version

# Output might look like:
# git version 2.44.0.windows.1
```

Just because Git is installed doesn't mean it tracks every file on your PC. You must explicitly tell Git **which folder** to track.

### The Scenario:
We have three project folders, but we only want to track `git_one`.

### The Command:

```bash

# 1. Navigate to the specific folder
$ cd git_one

# 2. Check current status (Should say "fatal: not a git repository")
$ git status

# 3. Initialize Git magic!
$ git init

# 4. Check status again
$ git status
```

### ğŸ•µï¸â€â™‚ï¸ The Secret `.git` Folder
When you run `git init`, Git creates a **hidden folder** named `.git`. This is the brain of your repository. It stores all the history, branches, and configurations.

**DO NOT TOUCH THIS FOLDER MANUALLY.** If you delete or mess with files inside `.git`, you will corrupt your project history.

To see it on Windows (Git Bash):

```bash
# List all files including hidden ones
$ ls -la

# Output:
# drwxr-xr-x 1 User 197121 0 Jan 13 19:42 .git
```

### 1. Working Directory ğŸ“
This is where you are right now. You create a file (e.g., `index.html`) and write code. Git knows the file is there, but it's not tracking it yet.

### 2. Staging Area (The Waiting Room) ğŸš‰
You run `git add`. You are telling Git, "Hey, I want to include this specific change in my next save point." It's like putting items in your shopping cart before checking out.

### 3. Commit (The Checkpoint) ğŸ’¾
You run `git commit`. Git takes everything in the Staging Area and freezes it in time. It gives it a unique ID (Hash). This is your permanent save point.
