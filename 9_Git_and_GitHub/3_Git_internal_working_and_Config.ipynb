# 1. Configuring Git (`git config`) ğŸ› ï¸

Git needs to know **who you are** before you can commit code. This information is burned into every commit you make.

### Global vs. Local
* **Global:** Applies to *every* project on your computer (Recommended for Name/Email).
* **Local:** Applies only to the *current* project.

### Setting Identity (Windows/Mac/Linux)

```bash

# Set your name (Use quotes if you have spaces)
$ git config --global user.name "Hitesh Choudhary"

# Set your email (Must match your GitHub email later)
$ git config --global user.email "hitesh@example.com"

# Verify the changes
$ git config --list
```

### âš¡ Pro Tip: Changing the Default Editor
By default, Git opens **Vim** for commit messages. Let's switch it to **VS Code**.

**Pre-requisite:** Open VS Code, press `Ctrl + Shift + P`, type "Install 'code' command in PATH".

```bash

# Tells Git to use VS Code and WAIT until you close the window
$ git config --global core.editor "code --wait"
```

# 3. Git Internals: How Commits Work ğŸ§¬

Git is essentially a **Linked List** of snapshots. Every commit knows who its parent is.

### The Anatomy of a Commit
1. **Hash (SHA-1):** A unique ID like `4c8a91...`.
2. **Author:** Who made it.
3. **Message:** What happened.
4. **Parent:** The pointer to the previous commit.

# 4. Exploring the `.git` Folder ğŸ•µï¸â€â™‚ï¸

Let's look inside the hidden folder where the magic happens.

**Command to peek:** `ls -la .git/`

| Folder/File | Purpose |
| :--- | :--- |
| `HEAD` | A pointer to your current branch (usually `refs/heads/master`). |
| `config` | Local configuration for *this* repo. |
| `hooks/` | Scripts that run automatically (e.g., `pre-commit`, `pre-push`). |
| `objects/` | Database of all your file contents and commits (compressed). |
| `refs/` | Pointers to branches and tags. |

### ğŸ§  One-Liners to Remember
* **.gitignore** is for secrets; if you commit a secret, it's in history forever!
* **SHA-1** hashes are unique; even changing a comma changes the hash completely.
* **HEAD** is just a pointer saying "You are here".
