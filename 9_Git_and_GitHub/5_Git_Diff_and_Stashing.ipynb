# 1. Git Diff: Spot the Difference ğŸ”

The `git diff` command compares two versions of code. It does NOT compare two different files (e.g., File A vs File B). It compares **File A (Version X)** vs **File A (Version Y)**.

### Reading the Symbols
Do NOT assume `-` means deleted and `+` means added in a simple way. It signifies **Version A** vs **Version B**.

| Symbol | Meaning |
| :--- | :--- |
| `---` | The content in the **first** source (e.g., committed version). |
| `+++` | The content in the **second** source (e.g., staged/working version). |

### Common Commands

| Command | Comparison |
| :--- | :--- |
| `git diff` | **Working Directory** vs **Staging Area** |
| `git diff --staged` | **Staging Area** vs **Last Commit** |
| `git diff <SHA1> <SHA2>` | **Commit 1** vs **Commit 2** |
| `git diff branch1..branch2` | **Branch 1** vs **Branch 2** |

```bash 

# Scenario: You modified index.html and STAGED it.
$ git add index.html

# Compare Staged changes vs Last Commit
$ git diff --staged

# Output Preview:
# --- a/index.html (Old Version)
# +++ b/index.html (New Version)
# - looks good as project
# + looks good project
```

# 2. Git Stash: The Temporary Shelf ğŸ“¦

### The Problem
You are working on `bug-fix` branch. The code is messy and broken. Suddenly, your boss asks you to switch to `footer` branch to help a colleague.
**Git won't let you switch** if you have uncommitted changes that conflict.

### The Solution: Stashing
Put your messy code on a temporary shelf (`stash`), switch branches, come back later, and retrieve it.

<pre style="white-space: pre; font-family: monospace; overflow-x: auto; color: #c9d1d9;">
   Working Directory              Stash (Shelf)
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”        â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚ Messy Code ğŸš§        â”€â”€â”€â”€â”€â–º       WIP: BugFix      
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜        â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
          â”‚                            â–²
      git stash                    git stash pop
</pre>

```bash
# 1. Save changes to stash (cleans your working directory)
$ git stash

# 2. Verify stash list
$ git stash list

# 3. Switch branches safely
$ git switch footer
# ... do your work ...
$ git switch bug-fix

# 4. Bring changes back and delete from stash
$ git stash pop

# OR bring changes back but KEEP in stash (copy)
$ git stash apply
```

> ğŸ’¡ **Tip:** Stashes are NOT branch-specific. You can stash changes on `branch-A` and pop them onto `branch-B`. Be careful!

# 3. Time Travel: Checking Out History â³

You can move your `HEAD` pointer to *any* commit in the past to see how the project looked at that moment.

### Detached HEAD State
When you check out a specific commit (not a branch), you enter **Detached HEAD** state. You can look around, but commits made here are easily lost unless you create a new branch.

### Methods to Time Travel

```bash
# Method 1: Using Commit Hash (SHA)
$ git log --oneline
# a1b2c3d Add Footer
# 9x8y7z6 Fix Bug

$ git checkout 9x8y7z6
# HEAD is now at 9x8y7z6... Fix Bug

# Method 2: Relative Movement (HEAD~n)
# Go back 2 commits from current position
$ git checkout HEAD~2

# ğŸš¨ HOW TO GO BACK TO PRESENT?
# Simply check out the branch name
$ git checkout master
```

# âš¡ Quick Revision & Cheat Sheet

| Command | Description | Memory Trick |
| :--- | :--- | :--- |
| `git diff --staged` | Review staged changes | The **"Final Check"** before committing |
| `git stash` | Save WIP changes temporarily | The **"Pause Button"** |
| `git stash pop` | Restore WIP changes | The **"Resume Button"** |
| `git checkout <SHA>` | View old code | The **"Time Machine"** |
| `git checkout master` | Return to present | **"Back to Reality"** |
