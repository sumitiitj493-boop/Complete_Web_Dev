# 1. What is Git Rebase? (Rewriting History) ğŸ“–

Rebasing is an alternative to Merging. It literally **rewrites history**.

### The Analogy
Imagine you branched off from Master yesterday. Since then, Master has moved forward. 
* **Merge:** Creates a bridge between your branch and the new Master (adds a "Merge Commit").
* **Rebase:** Lifts your entire branch up and **re-plants** it at the tip of the new Master. It looks like you started working *today* instead of yesterday.

### The Rule of Thumb
1. **Switch** to your side branch (e.g., `bug-fix`).
2. **Rebase** FROM the side branch onto the main branch.
3. **Never** run `git rebase` while you are sitting on `master`.

# 3. Practical Rebase Workflow ğŸ› ï¸

You are working on `bug-fix`. Master has updated. You want those updates in your branch without creating a "Merge Commit".

```bash

# 1. Ensure you are on YOUR branch (Not Master)
$ git checkout bug-fix

# 2. Run the rebase command
# "Hey Git, take my work and stack it on top of master"
$ git rebase master

# 3. If successful:
# Your history is now linear. You can now fast-forward merge into master.
```

# 4. Resolving Conflicts during Rebase âš”ï¸

This is where people panic. The workflow is **different** from a normal merge conflict.

**Scenario:** You run `git rebase master` and Git stops yelling about conflicts.

### ğŸ›‘ The Process

1. **Fix the Conflict:** Open VS Code, choose "Incoming" or "Current" changes.
2. **Add the changes:** `git add .`
3. **DO NOT COMMIT:** This is the most common mistake. You are in the *middle* of a process.
4. **Continue the Rebase:** Tell Git to proceed.

```bash
# 1. Fix file in VS Code

# 2. Stage the fix
$ git add index.html

# 3. Continue (Do NOT run git commit)
$ git rebase --continue

# 4. If things get too scary, EJECT!
$ git rebase --abort
```

# 5. Rebase vs. Merge: The Comparison âš–ï¸

| Feature | Git Merge | Git Rebase |
| :--- | :--- | :--- |
| **History** | Preserves truth (shows exactly when branches diverged). | Rewrites history (pretends work happened sequentially). |
| **Visual** | "Railroad tracks" or "Spaghetti" graph. | A clean, straight line. |
| **Safety** | Very Safe. Non-destructive. | **Dangerous.** Destructive (changes Commit IDs). |
| **Best For** | Bringing finished features into Master. | Updating your private branch with latest Master code. |

> ğŸ’¡ **Pro Tip:** Many companies force a "Linear History" policy, meaning you *must* rebase your feature branch before your Pull Request is accepted.
