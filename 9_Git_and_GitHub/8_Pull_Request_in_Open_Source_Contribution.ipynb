### Step 1: Forking ğŸ´
1. Go to the target repository on GitHub.
2. Click the **Fork** button (Top Right).
3. **Result:** You now have an exact copy of the project under *your* username (e.g., `yourname/project`).

### Step 2: Cloning ğŸ“¥
Bring *your* forked copy to your local machine.

```bash
# Clone YOUR version, not the original one
$ git clone [https://github.com/YOUR_USERNAME/project-name.git](https://github.com/YOUR_USERNAME/project-name.git)
$ cd project-name
```

### Creating a Feature Branch
Give your branch a descriptive name.

```bash
# Good branch names: navbar-fix, add-login, update-docs
$ git checkout -b navbar-feature
```

### Making Changes
1. Edit files (e.g., add `Navbar.html`).
2. Add changes.
3. Commit changes.

```bash
$ touch navbar.html
# ... write code ...
$ git add .
$ git commit -m "Add navigation bar feature"
```

# 5. Pushing to Your Fork ğŸš€

You cannot push to the original owner's repo. You push to **origin** (which is your fork).

```bash
# Push your SPECIFIC branch to your SPECIFIC origin
$ git push origin navbar-feature
```

**Output:** Git will upload the branch to your GitHub account and often gives you a URL to create a PR immediately.

# 6. The Pull Request (PR) ğŸ“¨

This is the moment of truth. You are asking the original owner to **Pull** your code into their repository.

### The UI Steps
1. Go to your GitHub repository (your fork).
2. You will see a yellow banner: **"navbar-feature had recent pushes... Compare & Pull Request"**.
3. Click that button.

### The PR Form (Critical!)
You will see a comparison:
* **Base Repository:** `original-owner/project` (main)
* **Head Repository:** `your-username/project` (navbar-feature)



### Writing the Description
**Do NOT leave this blank.** A blank PR is a rejected PR.
1. **Title:** Clear and concise (e.g., "Added Responsive Navbar").
2. **Description:**
    * What does this code do?
    * Why is it needed?
    * Screenshots (if UI related).
    * Mention the Issue Number (e.g., "Fixes #42").

```markdown
### Changes Made
- Added a new Navbar.html component.
- Linked Home, About, and Contact pages.

### Why
The previous index.html lacked navigation.
```

### Handling Feedback
1. **Don't Ghost:** If a maintainer asks for changes, reply.
2. **Be Patient:** Maintainers have jobs. They might take days or weeks to review.
3. **Draft PRs:** If your work isn't done but you want feedback, mark the PR as a "Draft".
