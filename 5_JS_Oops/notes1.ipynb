### Accessing & Setting Prototypes

```javascript
// Old way (Dunder Proto) â€” "Dunder" = Double Underscore
Lenovo.__proto__ = computer;

// Better way (Modern)
Object.setPrototypeOf(Tesla, genericCar);

// Getting prototype
Object.getPrototypeOf(Tesla);
```

## Prototype Chain Example

```javascript
// Generic car
let genericCar = {
    tires: 4
};

// Tesla inherits from genericCar
let Tesla = {
    driver: "AI"
};

// Set the prototype
Object.setPrototypeOf(Tesla, genericCar);

// Now Tesla has access to tires!
console.log(Tesla.driver);  // "AI" (own property)
console.log(Tesla.tires);   // 4 (inherited!)

// Get the inherited properties
console.log(Object.getPrototypeOf(Tesla)); // {tires: 4}
```

**Output:**
```
AI
4
{ tires: 4 }
```

# 5. Constructor Functions â€” The OG Way ğŸ”¨

## What is a Constructor Function?

A **constructor function** is a regular function that:
1. Starts with a **Capital Letter** (convention)
2. Uses the `this` keyword to assign properties
3. Is called with the `new` keyword

It's the **original way** to create object blueprints before ES6 classes!

## Constructor Function Syntax

```javascript
// Constructor Function (Capital P!)
function Person(name, age) {
    this.name = name;  // 'this' refers to new object
    this.age = age;
}

// Creating objects with 'new'
let john = new Person("John", 25);
let jane = new Person("Jane", 30);

console.log(john.name); // "John"
console.log(jane.age);  // 30
```

**Output:**
```
John
30
```

## Constructor with Methods

```javascript
function Tea(type) {
    this.type = type;
    
    // Method inside constructor
    this.describe = function() {
        return `This is a cup of ${this.type}`;
    };
}

let lemonTea = new Tea("Lemon Tea");
console.log(lemonTea.describe()); 

let masalaTea = new Tea("Masala Tea");
console.log(masalaTea.describe());
```

**Output:**
```
This is a cup of Lemon Tea
This is a cup of Masala Tea
```

## Adding Methods via Prototype

```javascript
// Define a constructor function
function Animal(type) {
    this.type = type;
}

// Add method via prototype (shared across all instances!)
Animal.prototype.speak = function() {
    return `${this.type} makes a sound!`;
};

// Create instances
let dog = new Animal("Dog");
let cat = new Animal("Cat");

console.log(dog.speak());
console.log(cat.speak());
```

**Output:**
```
Dog makes a sound!
Cat makes a sound!
```

# 6. ES6 Classes â€” Modern & Clean ğŸ’

ES6 introduced the `class` keyword as **syntactic sugar** over constructor functions. Same functionality, cleaner syntax!

## Class Components

| Component | Description |
|-----------|-------------|
| `constructor()` | Special method for initialization |
| Methods | Functions without `function` keyword |
| `static` | Methods belonging to class, not instances |
| `get`/`set` | Getters and setters |

## Class Syntax

```javascript
class Vehicle {
    // Constructor - called when using 'new'
    constructor(make, model) {
        this.make = make;
        this.model = model;
    }
    
    // Method (no 'function' keyword needed!)
    start() {
        return `${this.model} is a car from ${this.make}`;
    }
}

// Creating objects
let myCar = new Vehicle("Toyota", "Camry");
console.log(myCar.start());
```

**Output:**

# 7. Inheritance with `extends` ğŸ§¬

**Inheritance** allows a class to **inherit** properties and methods from another class.

### The Child Can:
- âœ… Access parent's properties and methods
- âœ… Add its own new properties and methods
- âœ… Override parent's methods (polymorphism!)

## Inheritance Example

```javascript
// Parent class
class Vehicle {
    constructor(make, model) {
        this.make = make;
        this.model = model;
    }
    
    start() {
        return `${this.model} from ${this.make} started!`;
    }
}

// Child class - extends Vehicle
class Car extends Vehicle {
    // Can add own methods
    drive() {
        return `${this.make} is driving!`;
    }
}

// Create Car object
let myCar = new Car("Toyota", "Corolla");
console.log(myCar.start());  // Inherited from Vehicle!
console.log(myCar.drive());  // Own method
```

**Output:**
```
Corolla from Toyota started!
Toyota is driving!
```

# 8. Encapsulation â€” Keep Your Secrets! ğŸ”

**Encapsulation** means **restricting direct access** to object data. We hide internal details and provide controlled access through methods.

**In JavaScript:** Use `#` prefix for **private** properties!

## Private Properties with #

```javascript
class BankAccount {
    #balance = 0;  // Private! Can't access outside
    
    deposit(amount) {
        this.#balance += amount;
        return this.#balance;
    }
    
    getBalance() {
        return `$${this.#balance}`;  // Controlled access
    }
}

let account = new BankAccount();
account.deposit(1000);
console.log(account.getBalance());
console.log(account.#balance);  // âŒ ERROR!
```

**Output:**
```
$1000
SyntaxError: Private field '#balance' must be declared in an enclosing class
```

# 10. Polymorphism â€” Many Forms! ğŸ­

**Polymorphism** (from Greek: "many forms") means the same method name can have **different behaviors** in different classes.

It's like the word "run" â€” a person runs, a program runs, water runs... same word, different actions!

## Polymorphism Example â€” Birds

```javascript
class Bird {
    fly() {
        return "Flying...";
    }
}

class Sparrow extends Bird {
    // Inherits fly() - can fly normally
}

class Penguin extends Bird {
    // Overrides fly() - different behavior!
    fly() {
        return "Penguins can't fly!";
    }
}

let sparrow = new Sparrow();
let penguin = new Penguin();

console.log(sparrow.fly());
console.log(penguin.fly());
// Same method name, different behaviors!
```

**Output:**
```
Flying...
Penguins can't fly!
```

# 11. Static Methods â€” Class-Level Powers! âš¡

**Static** members belong to the **class itself**, not to instances. They are called directly on the class, without creating an object!

Think of it like: "This is a factory feature, not a product feature."

## Static Methods Example

```javascript
class Calculator {
    // Static method - belongs to class
    static add(a, b) {
        return a + b;
    }
    
    // Regular method - belongs to instances
    multiply(a, b) {
        return a * b;
    }
}

// Static method - called on CLASS directly
console.log(Calculator.add(2, 3));

// Regular method - needs an instance
let calc = new Calculator();
console.log(calc.multiply(2, 3));

// This WON'T work!
console.log(calc.add(2, 3)); // ERROR!
```

**Output:**

# 12. Getters & Setters â€” Control Freaks! ğŸšï¸

**Getters** and **Setters** are special methods that let you **control** how properties are accessed (get) and modified (set). They look like properties but work like methods!

**Convention:** Use `_` prefix to indicate the property has a getter/setter.

## Getters & Setters Example

```javascript
class Employee {
    constructor(name, salary) {
        this.name = name;
        if (salary < 0) {
            throw new Error("Salary cannot be negative!");
        }
        this._salary = salary; // Convention: _ means has getter/setter
    }
    
    // Getter - called when accessing employee.salary
    get salary() {
        return `$${this._salary}`;
    }
    
    // Setter - called when setting employee.salary = value
    set salary(value) {
        if (value < 0) {
            console.error("Invalid salary!");
            return;
        }
        this._salary = value;
    }
}

let emp = new Employee("Alice", 50000);
console.log(emp.salary);     // Getter called
emp.salary = 60000;          // Setter called
console.log(emp.salary);     // Updated value
emp.salary = -1000;          // Invalid!
```

**Output:**
```
$50000
$60000
Invalid salary!
```
