### ğŸ”‘ SQL Key Concepts:

- Data **rows** aur **columns** mein store hota hai
- Exactly like Excel sheets!

- **Fixed structure** define karni padti hai pehle
- Agar name, email, age define kiya â†’ **Sabko** ye fields deni padegi
- Naya field add karna hai? **Poora table restructure!**

**3. Primary Key:**
- Har row ka **unique identifier**
- Jaise Aadhaar Card number - unique for every person!

**4. Foreign Key:**
- Ek table ko dusre table se **connect** karta hai
- "Ye Order kis User ka hai?" â†’ User_ID foreign key hai

**5. RDBMS:**
- Relational Database Management System
- Examples: MySQL, PostgreSQL, Oracle, SQLite
