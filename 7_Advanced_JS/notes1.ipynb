### ğŸ› ï¸ Creating a Promise â€“ The Syntax

```javascript
const fetchData = new Promise((resolve, reject) => {
    // Simulating a time-consuming operation
    setTimeout(() => {
        let success = true;  // Change to false to test reject
        
        if (success) {
            resolve("Data fetched successfully!");  // Everything good!
        } else {
            reject("Error fetching data");  // Something went wrong!
        }
    }, 3000);  // 3 seconds delay
});
```

### ğŸ¤ Interview Quick-Fire Answers

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">

**Q: What is a closure?**
> A: A closure is a function that remembers the environment in which it was created and retains access to variables from its outer scope even after the outer function has finished executing.

**Q: What is a Promise?**
> A: A Promise is an object representing the eventual completion or failure of an asynchronous operation. It has three states: pending, fulfilled, and rejected.

**Q: What does the Event Loop do?**
> A: The Event Loop continuously checks if the Call Stack is empty. If empty, it picks tasks from the queues (Microtask first, then Macrotask) and pushes them to the Call Stack for execution.

**Q: Why is JavaScript called single-threaded yet handles async operations?**
> A: JavaScript itself is single-threaded, but it delegates async operations (timers, network calls) to the browser/Node environment, which handles them separately and returns results via the callback queue.
