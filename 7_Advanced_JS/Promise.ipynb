# 3. Code Me Kaise Likhein? ğŸ› ï¸

Promise banane ke liye `new Promise()` use karte hain. Iske andar ek function hota hai jisme 2 cheezein milti hain: `resolve` (kaam ho gaya) aur `reject` (lag gayi).

```javascript

// Naya Promise banaya
const jaguarPromise = new Promise((resolve, reject) => {
    
    let paisaHaiKya = true; // Maan lo paise hain

    if (paisaHaiKya) {
        // Agar sab sahi hai toh RESOLVE call karo
        resolve("Le Baby, yeh rahi teri Jaguar! ğŸï¸");
    } else {
        // Agar gadbad hai toh REJECT call karo
        reject(new Error("Sorry Baby, Alto se kaam chala lo. ğŸ’”"));
    }
});
```

# 4. Promise Consume Karna (Result nikalna) ğŸ½ï¸

Promise toh bana liya, ab uska result kaise milega? Hum methods lagate hain.

| Method | Kab chalega? | Scene Kya Hai? |
| :--- | :--- | :--- |
| **`.then()`** | <span style="color:#3fb950">Jab Resolve ho</span> | Party time! Data mil gaya. |
| **`.catch()`** | <span style="color:#f85149">Jab Reject ho</span> | Error handling. Breakup ho gaya. |
| **`.finally()`** | <span style="color:#58a6ff">Hamesha</span> | Chahe kuch bhi ho, ye chalega (Cleanup). |

### Code Dekho:

```javascript
jaguarPromise
    .then((message) => {
        // Ye tab chalega jab resolve() call hoga
        console.log("Party On:", message);
    })
    .catch((error) => {
        // Ye tab chalega jab reject() call hoga
        console.error("Dhokha:", error.message);
    })
    .finally(() => {
        // Ye hamesha chalega
        console.log("Chalo ab so jao, bohot natak ho gaya.");
    });
```

# 6. Promise.all() - Gang wale kaam ğŸš¦

Maan le tere paas 3 kaam hain: 
1. Dost A ko pick karna hai.
2. Dost B ko pick karna hai.
3. Dost C ko pick karna hai.

Jab tak **TEENO** gadi me nahi baith jate, gadi nahi chalegi. Par agar **EK** ne bhi mana kar diya, toh trip cancel.

**Isi ko bolte hain `Promise.all()`**.

```javascript
// Teen alag-alag promises (dost)
const p1 = new Promise((resolve) => setTimeout(resolve, 1000, "Rahul Ready"));
const p2 = new Promise((resolve) => setTimeout(resolve, 2000, "Priya Ready"));
const p3 = new Promise((resolve) => setTimeout(resolve, 3000, "Amit Ready"));

// Wait karega jab tak teeno resolve na ho jayein
Promise.all([p1, p2, p3])
    .then((values) => {
        console.log("Sab aa gaye, chalo:", values);
    })
    .catch((err) => {
        console.log("Plan cancel bhai:", err);
    });
```
