<h2 style="color: #c9d1d9; margin:0;">ğŸ”— Part 1: Prototypal Inheritance</h2>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">ğŸ“š CONCEPT: Prototypal Inheritance</strong>
<p style="color: #c9d1d9;">In JavaScript, <strong>objects inherit properties from other objects via the prototype chain</strong>. Each object has a hidden <code>[[Prototype]]</code> property that refers to another object (its prototype). If a property can't be found on the object, JavaScript looks up the chain!</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #a371f7;">
<h3 style="color: #a371f7; margin-top:0;">ğŸ§  Memory Trick: It's Just Inheritance in Prototypes!</h3>
<p style="color: #c9d1d9;"><strong>Prototypal Inheritance = Inheritance being done in the prototypes.</strong> That's it! The whole concept is simply: you can inject methods/properties into a function's prototype, and all instances inherit them!</p>
</div>

### ğŸ¯ Interview-Ready Example

```javascript
// Step 1: Create a constructor function
function Person(name) {
    this.name = name;
}

// Step 2: Add method to prototype (THIS is prototypal inheritance!)
Person.prototype.greet = function() {
    console.log(`Hello, my name is ${this.name}`);
}

// Step 3: Create instance and use inherited method
let hitesh = new Person("Hitesh");
hitesh.greet();  // Output: Hello, my name is Hitesh
```

<h2 style="color: #c9d1d9; margin:0;">ğŸ‘† Part 2: The `this` Keyword & Context Binding</h2>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d; border-left: 5px solid #58a6ff;">
<strong style="color: #58a6ff;">ğŸ“š CONCEPT: The `this` Keyword</strong>
<p style="color: #c9d1d9;"><code>this</code> refers to the <strong>context</strong> in which a function is called. But here's the tricky part: when you transfer a method to another variable, <strong>the context can get LOST!</strong></p>
</div>

### ğŸ­ The Problem: Context Loss

```javascript
const person = {
    name: "Hitesh",
    greet: function() {
        console.log(`Hi, I am ${this.name}`);
    }
};

// This works perfectly!
person.greet();  // Output: Hi, I am Hitesh

// But what happens when we transfer the reference?
const greetFunction = person.greet;  // Just taking reference, NOT executing!

greetFunction();  // Output: Hi, I am undefined  ğŸ˜±
```

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #f85149; border-left: 5px solid #f85149;">
<strong style="color: #f85149;">ğŸ”´ WHY DOES THIS HAPPEN?</strong>
<p style="color: #c9d1d9;">When you transfer a method from one variable to another, <strong>the context is lost!</strong> The new variable doesn't know what <code>this</code> refers to anymore.</p>
</div>

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px dashed #a371f7;">
<h3 style="color: #a371f7; margin-top:0;">ğŸ§  Real-Life Analogy: The "Fan" Story</h3>
<p style="color: #c9d1d9;">Imagine you say: <em>"I got a fan!"</em> (meaning YouTube subscribers)</p>
<p style="color: #c9d1d9;">But when someone else repeats: <em>"He got a fan!"</em> â€“ the context is missing! They might think you sell ceiling fans! ğŸŒ€</p>
<p style="color: #8b949e;"><strong>Same with <code>this</code></strong> â€“ when you transfer the reference, the backstory (context) is lost!</p>
</div>

### ğŸ”§ The Solution: `bind()` Method

```javascript
const person = {
    name: "Hitesh",
    greet: function() {
        console.log(`Hi, I am ${this.name}`);
    }
};

// Without binding - context lost!
const greetFunction = person.greet;
greetFunction();  // Hi, I am undefined

// WITH binding - context preserved!
const boundGreet = person.greet.bind({ name: "John" });
boundGreet();  // Hi, I am John âœ…
```

### ğŸ¯ Creating a Promise (Same as Before)

```javascript
function fetchUserData() {
    return new Promise((resolve, reject) => {
        setTimeout(() => {
            // Simulating data fetch after 3 seconds
            resolve({
                name: "chaicode",
                url: "chaicode.com"
            });
        }, 3000);
    });
}
```

### ğŸ”„ Consuming with Async/Await

```javascript
async function getUserData() {
    try {
        console.log("Fetching user data...");
        
        const userData = await fetchUserData();  // WAITS here!
        
        console.log("User data fetched successfully!");
        console.log("User data:", userData);
        
    } catch (error) {
        console.log("Error fetching data");
        throw error;
    }
}

getUserData();  // Don't forget to call it!
```

### ğŸ“ The Scenario: Multiple Data Sources

```javascript
function fetchPostData() {
    return new Promise((resolve) => {
        setTimeout(() => {
            resolve("Post data fetched");
        }, 2000);  // 2 seconds
    });
}

function fetchCommentData() {
    return new Promise((resolve) => {
        setTimeout(() => {
            resolve("Comment data fetched");
        }, 3000);  // 3 seconds
    });
}
```

### ğŸ¢ Method 1: Sequential Awaits (Works, but slower)

```javascript
async function getBlogData() {
    try {
        console.log("Fetching blog data...");
        
        const blogData = await fetchPostData();     // Wait 2 sec
        const commentData = await fetchCommentData(); // Then wait 3 sec
        
        console.log(blogData);
        console.log(commentData);
        console.log("Fetch complete!");
        
    } catch (error) {
        console.log("Error fetching blog data");
    }
}
// Total time: 2 + 3 = 5 seconds (sequential!)
```

### ğŸš€ Method 2: Promise.all (Better! Parallel execution)

```javascript
async function getBlogData() {
    try {
        console.log("Fetching blog data...");
        
        const [postData, commentData] = await Promise.all([
            fetchPostData(),
            fetchCommentData()
        ]);
        
        console.log(postData);
        console.log(commentData);
        console.log("Fetch complete!");
        
    } catch (error) {
        console.log("Error fetching blog data");
    }
}
// Total time: max(2, 3) = 3 seconds (parallel!)
```

### ğŸ¯ Generator Syntax

```javascript
// Note the asterisk (*) after function â€“ this makes it a generator!
function* numberGenerator() {
    yield 1;  // First call returns this
    yield 2;  // Second call returns this
    yield 3;  // Third call returns this
    // After this, returns undefined
}

const gen = numberGenerator();

console.log(gen);  // Object [Generator] {}

console.log(gen.next().value);  // 1
console.log(gen.next().value);  // 2
console.log(gen.next().value);  // 3
console.log(gen.next().value);  // undefined (no more yields!)
```

### ğŸ¤ Interview Quick-Fire Answers

<div style="background-color: #161b22; padding: 15px; border-radius: 10px; border: 1px solid #30363d;">

**Q: What is prototypal inheritance?**
> A: In JavaScript, objects inherit properties from other objects via the prototype chain. Each object has a hidden `[[Prototype]]` property linking to its prototype. You can add methods to `Constructor.prototype` and all instances will inherit them.

**Q: Why does `this` become undefined sometimes?**
> A: When you transfer a method reference to another variable, the context (what `this` refers to) is lost. The new variable doesn't know the original object. Use `bind()` to preserve context.

**Q: What's the difference between bind, call, and apply?**
> A: `bind()` creates a new function with bound context (doesn't execute). `call()` and `apply()` execute immediately with specified context. Difference: `call` takes arguments individually, `apply` takes them as an array.

**Q: What's async/await?**
> A: A modern way to consume promises that makes async code look synchronous. `async` marks a function, `await` pauses execution until a promise resolves. Must use try/catch for error handling.

**Q: What's a generator function?**
> A: A function (declared with `function*`) that can pause and resume execution. Uses `yield` instead of `return`. Calling `.next()` gets the next yielded value. Rarely used in application code.

**Q: ES6 Modules vs CommonJS?**
> A: ES6 uses `export`/`import` syntax (modern standard). CommonJS uses `module.exports`/`require()` (older, still in npm packages). Use ES6 for new projects!
