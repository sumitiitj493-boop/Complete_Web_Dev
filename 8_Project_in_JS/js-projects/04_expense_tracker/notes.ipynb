# 4. Form Submission & Validation ğŸ“

The heart of our application - where user input becomes structured data.

```javascript
expenseForm.addEventListener('submit', (e) => {
  // CRITICAL: Prevent page reload!
  e.preventDefault();
  
  // Get input values
  const name = expenseNameInput.value.trim();
  const amount = parseFloat(expenseAmountInput.value.trim());
  
  // Triple validation
  if (name !== '' && !isNaN(amount) && amount > 0) {
    // Create expense object
    const newExpense = {
      id: Date.now(),
      name: name,      // or just: name,
      amount: amount   // or just: amount
    };
    
    // Add to array
    expenses.push(newExpense);
    
    // Save, render, update
    saveExpensesToLocal();
    renderExpenses();
    updateTotal();
    clearInputs();
  }
});
```

# 6. Array.reduce() Method ğŸ”¢

The professional way to calculate totals from arrays.

### Basic Reduce Example

```javascript
const numbers = [1, 2, 3, 4];

const sum = numbers.reduce((accumulator, currentValue) => {
  return accumulator + currentValue;
}, 0);  // 0 is the initial value

console.log(sum);  // 10
```

**How it works:**

<pre style="white-space: pre; font-family: monospace; overflow-x: auto; background: #161b22; padding: 20px; border-radius: 8px; border: 1px solid #30363d;">
Step 1: accumulator = 0, currentValue = 1 â†’ return 0 + 1 = 1
Step 2: accumulator = 1, currentValue = 2 â†’ return 1 + 2 = 3
Step 3: accumulator = 3, currentValue = 3 â†’ return 3 + 3 = 6
Step 4: accumulator = 6, currentValue = 4 â†’ return 6 + 4 = 10

Final result: 10
</pre>

### Calculate Expense Total

```javascript
function calculateTotal() {
  return expenses.reduce((sum, expense) => {
    return sum + expense.amount;
  }, 0);
  
  // Shorter syntax
  return expenses.reduce((sum, expense) => sum + expense.amount, 0);
}
```

**With expense objects:**

```javascript
expenses = [
  {id: 1, name: 'Movie', amount: 100},
  {id: 2, name: 'Course', amount: 5}
]

calculateTotal();
// Step 1: sum=0, expense={...Movie}, return 0 + 100 = 100
// Step 2: sum=100, expense={...Course}, return 100 + 5 = 105
// Result: 105
```

### Update Total Function

```javascript
function updateTotal() {
  const totalAmount = calculateTotal();
  totalAmountDisplay.textContent = totalAmount.toFixed(2);
}

// toFixed(2) ensures 2 decimal places
// 105 â†’ "105.00"
// 105.5 â†’ "105.50"
// 105.567 â†’ "105.57"
```

### Reduce Parameters

| Parameter | Description | Common Names |
|-----------|-------------|---------------|
| accumulator | Running total | `sum`, `acc`, `total` |
| currentValue | Current item | `item`, `curr`, `expense` |
| index (optional) | Current position | `i`, `index` |
| array (optional) | Original array | `arr`, `array` |
| initialValue | Starting value | Usually `0` for sums |

> ğŸ’¡ **Tip:** `reduce()` is powerful but can be confusing. Think of the accumulator as a bucket that collects values as you loop!

# 7. Rendering Expenses ğŸ¨

Transform array data into visible HTML elements.

```javascript
function renderExpenses() {
  // Step 1: Clear existing list
  expenseList.innerHTML = '';
  
  // Step 2: Loop through expenses
  expenses.forEach((expense) => {
    // Step 3: Create list item
    const li = document.createElement('li');
    
    // Step 4: Set content
    li.innerHTML = `
      ${expense.name} - $${expense.amount.toFixed(2)}
      <button data-id="${expense.id}">Delete</button>
    `;
    
    // Step 5: Append to list
    expenseList.appendChild(li);
  });
}
```

# 8. Delete Functionality ğŸ—‘ï¸

Remove items from array, update storage, and refresh display.

```javascript
expenseList.addEventListener('click', (e) => {
  // Check if clicked element is a button
  if (e.target.tagName === 'BUTTON') {
    
    // Get expense ID from data-id
    const expenseId = parseInt(e.target.getAttribute('data-id'));
    
    // Filter out the matching expense
    expenses = expenses.filter(expense => expense.id !== expenseId);
    
    // Save, render, update
    saveExpensesToLocal();
    renderExpenses();
    updateTotal();
  }
});
```

### How Filter Works

```javascript
expenses = [
  {id: 1, name: 'Movie', amount: 100},
  {id: 2, name: 'Course', amount: 5},
  {id: 3, name: 'Food', amount: 20}
];

// Delete expense with id=2
const expenseId = 2;
expenses = expenses.filter(expense => expense.id !== expenseId);

// Filter keeps items where condition is TRUE:
// id: 1 !== 2 â†’ true (keep Movie)
// id: 2 !== 2 â†’ false (REMOVE Course)
// id: 3 !== 2 â†’ true (keep Food)

// Result: [{id:1,...}, {id:3,...}]
```

# 10. Complete Code ğŸ“

Here's the full, production-ready expense tracker:

```javascript
document.addEventListener('DOMContentLoaded', () => {
  
  // DOM References
  const expenseForm = document.getElementById('expense-form');
  const expenseNameInput = document.getElementById('expense-name');
  const expenseAmountInput = document.getElementById('expense-amount');
  const expenseList = document.getElementById('expense-list');
  const totalAmountDisplay = document.getElementById('total-amount');
  
  // Data
  let expenses = [];
  
  // Load from localStorage
  const savedExpenses = localStorage.getItem('expenses');
  if (savedExpenses) {
    expenses = JSON.parse(savedExpenses);
    renderExpenses();
    updateTotal();
  }
  
  // Form Submit
  expenseForm.addEventListener('submit', (e) => {
    e.preventDefault();
    
    const name = expenseNameInput.value.trim();
    const amount = parseFloat(expenseAmountInput.value.trim());
    
    if (name !== '' && !isNaN(amount) && amount > 0) {
      const newExpense = {
        id: Date.now(),
        name,
        amount
      };
      
      expenses.push(newExpense);
      saveExpensesToLocal();
      renderExpenses();
      updateTotal();
      clearInputs();
    }
  });
  
  // Delete Handler
  expenseList.addEventListener('click', (e) => {
    if (e.target.tagName === 'BUTTON') {
      const expenseId = parseInt(e.target.getAttribute('data-id'));
      expenses = expenses.filter(expense => expense.id !== expenseId);
      saveExpensesToLocal();
      renderExpenses();
      updateTotal();
    }
  });
  
  // Functions
  function saveExpensesToLocal() {
    localStorage.setItem('expenses', JSON.stringify(expenses));
  }
  
  function renderExpenses() {
    expenseList.innerHTML = '';
    expenses.forEach((expense) => {
      const li = document.createElement('li');
      li.innerHTML = `
        ${expense.name} - $${expense.amount.toFixed(2)}
        <button data-id="${expense.id}">Delete</button>
      `;
      expenseList.appendChild(li);
    });
  }
  
  function calculateTotal() {
    return expenses.reduce((sum, expense) => sum + expense.amount, 0);
  }
  
  function updateTotal() {
    const totalAmount = calculateTotal();
    totalAmountDisplay.textContent = totalAmount.toFixed(2);
  }
  
  function clearInputs() {
    expenseNameInput.value = '';
    expenseAmountInput.value = '';
    expenseNameInput.focus();
  }
  
});
```

### Memory Tricks ğŸ§ 

**The 3 V's of Validation:**
1. **V**alue exists (not empty)
2. **V**alid number (not NaN)
3. **V**iable amount (greater than 0)

**The Storage Sandwich:**
- **Bread 1:** JSON.stringify (going in)
- **Filling:** LocalStorage
- **Bread 2:** JSON.parse (coming out)

**The CRUD Cycle:**
- **C**reate â†’ push()
- **R**ead â†’ forEach(), render()
- **U**pdate â†’ modify then save
- **D**elete â†’ filter()

**Reduce Mantra:**
- *"Accumulate like a pro - start at zero, add each item, return the total"*
