# 3. Conditional Styling (Completed Tasks) ğŸ¨

If a task is completed, we want to show it with a **strike-through**. We do this by adding a CSS class conditionally.

```js
function renderTask(task) {
  const li = document.createElement('li');
  li.setAttribute('data-id', task.id);
  
  li.innerHTML = `
    <span>${task.text}</span>
    <button>Delete</button>
  `;
  
  // CONDITIONAL: Add 'completed' class if task is done
  if (task.completed) {
    li.classList.add('completed');
  }
  
  todoList.appendChild(li);
}
```

# 4. Toggle Task Completion ğŸ”„

When the user clicks on a task (but NOT the delete button), we want to:

1. Toggle the `completed` property (true â†” false)
2. Toggle the CSS class for visual feedback
3. Save to Local Storage

```js
// Add event listener to the <li> element
li.addEventListener('click', (event) => {
  
  // GUARD: If click came from button, ignore
  if (event.target.tagName === 'BUTTON') {
    return;  // Stop execution here!
  }
  
  // Toggle the completed status
  task.completed = !task.completed;
  
  // Toggle the CSS class
  li.classList.toggle('completed');
  
  // Save changes to Local Storage
  saveTasks();
});
```

<div style='background: #161b22; padding: 18px; border-left: 4px solid #d29922; border-radius: 6px;'>

### ğŸ§  The NOT Operator Magic: `!`

The exclamation mark `!` is called the **NOT operator**. It flips boolean values:

| Before | Operation | After |
|--------|-----------|-------|
| `true` | `!true` | `false` |
| `false` | `!false` | `true` |

So `task.completed = !task.completed` is a **one-liner toggle**!

</div>

<div style='background: #161b22; padding: 18px; border-left: 4px solid #f85149; border-radius: 6px;'>

### âš ï¸ Why Check `tagName === 'BUTTON'`?

The `<li>` contains both the task text AND the delete button.

If we don't check, clicking "Delete" would **also** trigger the toggle! We use a **Guard Clause** to prevent this.

**Important Note:** `tagName` is always UPPERCASE (`'BUTTON'`, not `'button'`)

## Deleting a Task with `filter()`

```js
li.querySelector('button').addEventListener('click', (e) => {
  e.stopPropagation(); // Prevent toggle from firing
  
  // STEP 1: Remove from array
  // Keep all tasks EXCEPT the one with matching ID
  tasks = tasks.filter(t => t.id !== task.id);
  
  // STEP 2: Remove from DOM
  li.remove();
  
  // STEP 3: Save to Local Storage
  saveTasks();
});
```

```html
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚  <li>                                          â”‚
â”‚                                                â”‚
â”‚    â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”  â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”   â”‚
â”‚    â”‚<span>Task Text</span>â”‚  â”‚  <button>   â”‚   â”‚ â† Target!
â”‚    â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜  â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜   â”‚
â”‚                                                â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
                    â†‘
                    â”‚
        li.querySelector('button')
        (searches ONLY inside this li)


<div style='background: #161b22; padding: 18px; border-left: 4px solid #3fb950; border-radius: 6px;'>

### ğŸ’¡ querySelector vs getElementById

| Method | Syntax | Use Case |
|--------|--------|----------|
| `getElementById` | `('myId')` | Fast, but ID only |
| `querySelector` | `('.class')` or `('#id')` or `('tag')` | Flexible, CSS-like |
| `querySelectorAll` | Same as above | Returns ALL matches |

**Pro Tip:** Most developers prefer `querySelector` for its flexibility!

</div>

---

# 8. Re-Rendering: Keeping UI in Sync ğŸ”„

After adding a new task, the data is saved but the UI doesn't update automatically.

**Solution:** Call `renderTask()` immediately after adding!

```js
addTaskBtn.addEventListener('click', () => {
  const value = todoInput.value.trim();
  if (value === "") return;
  
  const newTask = {
    id: Date.now(),
    text: value,
    completed: false
  };
  
  tasks.push(newTask);     // Update array
  saveTasks();             // Save to storage
  renderTask(newTask);     // IMMEDIATELY render!
  todoInput.value = "";    // Clear input
});
```

# 9. The Complete Code ğŸ’»

```js
// ========== DOM SELECTION ==========
const todoInput = document.getElementById('todo-input');
const addTaskBtn = document.getElementById('add-task-btn');
const todoList = document.getElementById('todo-list');

let tasks = [];

// ========== SAVE TO LOCAL STORAGE ==========
function saveTasks() {
  localStorage.setItem('tasks', JSON.stringify(tasks));
}

// ========== RENDER SINGLE TASK ==========
function renderTask(task) {
  const li = document.createElement('li');
  li.setAttribute('data-id', task.id);
  
  li.innerHTML = `
    <span>${task.text}</span>
    <button>Delete</button>
  `;
  
  if (task.completed) {
    li.classList.add('completed');
  }
  
  // Toggle complete on li click
  li.addEventListener('click', (e) => {
    if (e.target.tagName === 'BUTTON') return;
    task.completed = !task.completed;
    li.classList.toggle('completed');
    saveTasks();
  });
  
  // Delete on button click
  li.querySelector('button').addEventListener('click', (e) => {
    e.stopPropagation();
    tasks = tasks.filter(t => t.id !== task.id);
    li.remove();
    saveTasks();
  });
  
  todoList.appendChild(li);
}

// ========== LOAD ON PAGE LOAD ==========
document.addEventListener('DOMContentLoaded', () => {
  tasks = JSON.parse(localStorage.getItem('tasks')) || [];
  tasks.forEach(task => renderTask(task));
});

// ========== ADD NEW TASK ==========
addTaskBtn.addEventListener('click', () => {
  const value = todoInput.value.trim();
  if (value === '') return;
  
  const newTask = {
    id: Date.now(),
    text: value,
    completed: false
  };
  
  tasks.push(newTask);
  saveTasks();
  renderTask(newTask);
  todoInput.value = '';
});
```
