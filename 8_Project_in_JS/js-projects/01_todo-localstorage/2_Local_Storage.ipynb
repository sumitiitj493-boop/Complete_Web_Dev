## 2. Implementing `saveTasks()` ğŸ“

We create a dedicated function to handle saving. We call this function every time we **Add**, **Delete**, or **Update** a task.

```javascript
function saveTasks() {
    // 1. Take the array 'tasks'
    // 2. Convert to String: JSON.stringify(tasks)
    // 3. Save to Storage: setItem('key', 'value')
    
    localStorage.setItem("tasks", JSON.stringify(tasks));
}
```

> **ğŸ’¡ Pro Tip:** You can verify this in Chrome DevTools â†’ **Application** tab â†’ **Local Storage**.

## 4. The `DOMContentLoaded` Event ğŸ›¡ï¸

We can't just run code immediately. We must ensure the HTML (DOM) is fully loaded before we try to display tasks.

### The Initialization Logic
We use the **Logical OR (`||`)** operator to handle the case where the user visits for the very first time (and storage is empty).

```javascript
document.addEventListener('DOMContentLoaded', () => {
    // Step 1: Get data from storage
    // If tasks exist? Parse them. 
    // If null? Use empty array [] to avoid errors.
    tasks = JSON.parse(localStorage.getItem("tasks")) || [];

    // Step 2: Render existing tasks
    tasks.forEach((task) => {
        renderTask(task); // We will build this function next video
    });
});
```
