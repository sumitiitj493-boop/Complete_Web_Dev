## 5. Step 3: Writing the Code (Event Listener) âŒ¨ï¸

We use `addEventListener` on the button.

```javascript
addTaskBtn.addEventListener('click', () => {
    
    // A. Get the value & clean it
    const value = todoInput.value.trim(); 
    
    // B. Validation (Guard Clause)
    if(value === "") {
        return; // Stop execution immediately!
    }

    // C. Create the Task Object
    const newTask = {
        id: Date.now(),      // Unique ID based on time
        text: value,         // The actual task
        completed: false     // Default status
    };

    // D. Update Data
    tasks.push(newTask);

    // E. Reset UI
    todoInput.value = "";

    console.log(tasks); // Debugging check
});
```

## 6. Deep Dive: Key Concepts ğŸ¤¿

### A. The `.trim()` method
Why use it? If a user types `"   Gym   "`, we only want `"Gym"`. Also, if they type `"   "` (just spaces), `trim()` turns it into `""` (empty string), so our check catches it.

### B. `Date.now()` for IDs
We need a unique ID to find/delete items later. `Date.now()` returns the number of milliseconds since 1970. It changes every millisecond, making it a great simple unique ID generator.

### C. The Debugging Mistake
