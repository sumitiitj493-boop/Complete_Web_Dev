```javascript
// Command ke according action lo
if (command === 'add') {
    addTask(argument);      // Task add karo
} 
else if (command === 'list') {
    listTasks();            // Tasks list karo
} 
else if (command === 'remove') {
    removeTask(parseInt(argument));  // Task remove karo
    // parseInt kyunki terminal se string aata hai!
} 
else {
    console.log('Command not recognized!');
}
```

## ğŸ“¦ Complete Code - Pura Project!

```javascript
// ============================================
// TO-DO CLI APPLICATION - todo.js
// ============================================

// Step 1: FS Module import karo
const fs = require('fs');

// Step 2: File path define karo
const filePath = './tasks.json';

// Step 3: Command aur Argument grab karo
const command = process.argv[2];
const argument = process.argv[3];

// ============================================
// HELPER FUNCTIONS
// ============================================

// Tasks load karne ka function
const loadTasks = () => {
    try {
        const dataBuffer = fs.readFileSync(filePath);
        const dataJSON = dataBuffer.toString();
        return JSON.parse(dataJSON);
    } catch (error) {
        return [];
    }
};

// Tasks save karne ka function
const saveTasks = (tasks) => {
    const dataJSON = JSON.stringify(tasks);
    fs.writeFileSync(filePath, dataJSON);
};

// ============================================
// MAIN FUNCTIONS
// ============================================

// Task add karo
const addTask = (task) => {
    const tasks = loadTasks();
    tasks.push({ task: task });
    saveTasks(tasks);
    console.log(`Task added: ${task}`);
};

// Tasks list karo
const listTasks = () => {
    const tasks = loadTasks();
    tasks.forEach((task, index) => {
        console.log(`${index + 1} - ${task.task}`);
    });
};

// Task remove karo
const removeTask = (id) => {
    const tasks = loadTasks();
    tasks.splice(id - 1, 1);  // id - 1 because array is 0-indexed
    saveTasks(tasks);
    console.log(`Task ${id} removed!`);
};

// ============================================
// COMMAND HANDLER
// ============================================

if (command === 'add') {
    addTask(argument);
} else if (command === 'list') {
    listTasks();
} else if (command === 'remove') {
    removeTask(parseInt(argument));
} else {
    console.log('Command not recognized!');
}
```
