## ğŸ“¦ Complete Code - Full Restaurant Setup!

</div>

```javascript
// ============================================
// APNA KHUD KA NGINX - server.js
// ============================================

// Step 1: Teen Musketeers bulao
const http = require('http');
const fs = require('fs');
const path = require('path');

// Step 2: Port fix karo
const PORT = 3000;

// Step 3: MIME Types define karo (Menu Categories)
const mimeTypes = {
    '.html': 'text/html',
    '.css': 'text/css',
    '.js': 'application/javascript',
    '.png': 'image/png',
    '.jpg': 'image/jpeg',
    '.json': 'application/json'
};

// Step 4: Server banao (Waiter hire karo)
const server = http.createServer((request, response) => {
    
    // Step 5: File path dhundho
    const filePath = path.join(
        __dirname,
        request.url === '/' ? 'index.html' : request.url
    );
    
    // Debugging ke liye (optional)
    console.log(`Request aaya: ${request.url}`);
    console.log(`File path: ${filePath}`);
    
    // Step 6: Extension nikalo
    const extName = String(path.extname(filePath)).toLowerCase();
    
    // Step 7: Content type set karo
    const contentType = mimeTypes[extName] || 'application/octet-stream';
    
    // Step 8: File padho aur serve karo
    fs.readFile(filePath, (error, content) => {
        
        if (error) {
            // File nahi mili - 404 bhejo
            if (error.code === 'ENOENT') {
                response.writeHead(404, { 'Content-Type': 'text/html' });
                response.end('404: File Not Found Bro!', 'utf-8');
            }
        } else {
            // Sab badhiya - file serve karo
            response.writeHead(200, { 'Content-Type': contentType });
            response.end(content, 'utf-8');
        }
    });
});

// Step 9: Restaurant kholo!
server.listen(PORT, () => {
    console.log(`Server is listening on port ${PORT}`);
    console.log(`Visit: http://localhost:${PORT}`);
});
```
