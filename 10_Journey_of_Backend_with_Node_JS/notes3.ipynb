### ğŸ« Flashback #1: Classes Aur Inheritance Kya Hoti Hai?

<div style="background-color: #161b22; padding: 20px; color: #c9d1d9; border-radius: 10px; margin: 15px 0;">

**Socho ek Family ka scene:**

**Baap (Parent Class):** Sharma Ji ğŸ‘¨
- Properties: Surname "Sharma", Ghar ka address
- Methods: Kaam karna, Khana khana

**Beta (Child Class):** Rahul Sharma ğŸ‘¦
- **Inherited:** Surname "Sharma", Ghar ka address, Khana khana

**Inheritance = Baap ki properties + Apni special powers!**

```javascript
// Parent Class
class Animal {
    eat() {
        console.log("Khana kha raha hoon");
    }
}

// Child Class - extends keyword se inherit karo
class Dog extends Animal {
    bark() {
        console.log("Bhow Bhow!");  // Apna special method
    }
}

const tommy = new Dog();
tommy.eat();   // Inherited from Animal
tommy.bark();  // Dog ka apna
```

### ğŸ¨ Custom EventEmitter Class Banana

<div style="background-color: #161b22; padding: 20px; color: #c9d1d9; border-radius: 10px; margin: 15px 0;">

**Ab hum EventEmitter ko extend karke apni CUSTOM class banayenge!**

**Kyun?** Kyunki hum chahte hain:
1. EventEmitter ki **saari powers** (emit, on, once...)
2. Plus **apne custom methods** (log, etc.)

**Inheritance ka asli use!** ğŸ’ª

</div>

```javascript
const EventEmitter = require('events');

// Custom Logger class jo EventEmitter se inherit karti hai
class Logger extends EventEmitter {
    // Apna custom method
    log(message) {
        // Inherited emit method use karo
        this.emit('message', { message });
    }
}
```

### Step 1: Modules Import Karo

```javascript
// Required modules import karo
const fs = require('fs');           // File mein likhne ke liye
const os = require('os');           // System info ke liye
const EventEmitter = require('events');  // Events ke liye (CLASS hai!)
```

<div style="background-color: #1c1c1c; border-left: 5px solid #58a6ff; padding: 15px; color: #c9d1d9; margin: 15px 0; border-radius: 0 10px 10px 0;">

**ğŸ’¡ Notice kiya?**

`EventEmitter` ko **Capital letter** se likha - kyunki ye ek **CLASS** hai, regular module nahi!

Classes ko Capital se likhna convention hai (PascalCase).

</div>

### Step 2: Custom Logger Class Banao

```javascript
// Custom Logger class - EventEmitter se inherit karti hai
class Logger extends EventEmitter {
    // Custom log method
    log(message) {
        // "message" naam ka event emit karo
        // Data ke saath { message: message } object bhejo
        this.emit('message', { message });
    }
}
```

<div style="background-color: #161b22; padding: 20px; color: #c9d1d9; border-radius: 10px; margin: 15px 0;">

**Kya ho raha hai yahan?**

1. `class Logger extends EventEmitter` = Logger ko EventEmitter ki saari powers mil gayi!
2. `log(message)` = Apna custom method
3. `this.emit('message', { message })` = Event broadcast karo
   - `'message'` = Event ka naam (Frequency!)
   - `{ message }` = Data jo bhej rahe ho

</div>

### Step 3: Logger Object aur File Path

```javascript
// Logger ka object banao
const logger = new Logger();

// Log file ka path
const logFile = './event.log';  // Ya event.txt
```

<div style="background-color: #1c1c1c; border-left: 5px solid #3fb950; padding: 15px; color: #c9d1d9; margin: 15px 0; border-radius: 0 10px 10px 0;">

**ğŸ’¡ .log vs .txt:**

Dono same hain! `.log` sirf convention hai production mein.

Datadog, Sentry jaise tools `.log` files prefer karte hain.

### Step 4: File Mein Log Karne Ka Function

```javascript
// File mein log karne ka function
const logToFile = (event) => {
    // Log message craft karo with timestamp
    const logMessage = `${new Date().toISOString()} - ${event.message}\n`;
    
    // File mein append karo (add at bottom)
    fs.appendFileSync(logFile, logMessage);
};
```

<div style="background-color: #161b22; padding: 20px; color: #c9d1d9; border-radius: 10px; margin: 15px 0;">

**Breakdown:**

| Part | Kya Karta Hai |
|------|---------------|
| `new Date()` | Current date/time |
| `.toISOString()` | International format mein convert |
| `event.message` | Jo message emit hua |
| `\n` | New line (neeche likhne ke liye) |
| `appendFileSync` | File ke end mein add karo |

**appendFile vs writeFile:**
- `writeFile` = Purana data mita do, naya likho
- `appendFile` = Purana rehne do, neeche add karo âœ…

</div>

### Step 5: Event Listener Register Karo

```javascript
// Listener register karo - "message" event ke liye
logger.on('message', logToFile);
```

### Step 6: Periodic Memory Check with setInterval

```javascript
// Initial log
logger.log('Application started');

// Har 3 second mein memory check karo
setInterval(() => {
    // Free memory percentage calculate karo
    const freeMemory = os.freemem();
    const totalMemory = os.totalmem();
    const memoryUsage = ((freeMemory / totalMemory) * 100).toFixed(2);
    
    // Event emit karo with memory info
    logger.log(`Current memory usage: ${memoryUsage}% free`);
    
}, 3000);  // 3000ms = 3 seconds
```

## ğŸ“¦ Complete Code - Full Project!

```javascript
// ============================================
// CUSTOM EVENT LOGGER - logger.js
// ============================================

// Step 1: Required modules import karo
const fs = require('fs');
const os = require('os');
const EventEmitter = require('events');

// ============================================
// Step 2: Custom Logger Class (EventEmitter extend)
// ============================================
class Logger extends EventEmitter {
    log(message) {
        // "message" event emit karo with data
        this.emit('message', { message });
    }
}

// ============================================
// Step 3: Logger object aur file path
// ============================================
const logger = new Logger();
const logFile = './event.log';

// ============================================
// Step 4: File mein log karne ka function
// ============================================
const logToFile = (event) => {
    // Timestamp + message format
    const logMessage = `${new Date().toISOString()} - ${event.message}\n`;
    
    // File mein append karo
    fs.appendFileSync(logFile, logMessage);
};

// ============================================
// Step 5: Listener register karo
// ============================================
logger.on('message', logToFile);

// ============================================
// Step 6: Application start log
// ============================================
logger.log('Application started');
console.log('Logger started! Check event.log file.');

// ============================================
// Step 7: Har 3 second mein memory check
// ============================================
setInterval(() => {
    // Memory calculate karo
    const freeMemory = os.freemem();
    const totalMemory = os.totalmem();
    const memoryUsage = ((freeMemory / totalMemory) * 100).toFixed(2);
    
    // Log the event
    logger.log(`Memory Status: ${memoryUsage}% free`);
    
    console.log(`Logged: ${memoryUsage}% memory free`);
    
}, 3000);
```
